```python
import torch
from torch.utils.data import Dataset, DataLoader
from pytorch_nnue.data_loader import HalfKPDataset
import time
```

```python
dataset = HalfKPDataset(batch_size=1024*8, shuffle=True)

dataloader = DataLoader(
        dataset,
        batch_size=None,          # Le batching est désormais géré nativement (sans overhead) par le Dataset
        num_workers=6,
        pin_memory=True,          # Extrèmement important pour le transfert rapide vers le GPU
        persistent_workers=True,
        prefetch_factor=4,
    )
    
print("Début du benchmarking de l'IterableDataset vectorisé...")
start = time.perf_counter()
times = []
for i, batch in enumerate(dataloader):
    stm_indices, nstm_kings, (score, wdl) = batch
    
    elapsed = time.perf_counter() - start
    times.append(elapsed)
    print(f"i: {i}, batch size: {stm_indices.shape[0]}, elapsed time: {elapsed:.4f} s")
    start = time.perf_counter()
    if i > 1000:
        break
print(f"average time: {max(0, sum(times) / len(times)):.4f} s")
```
